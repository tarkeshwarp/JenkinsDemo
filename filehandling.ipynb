```
# Recurrsion 

def sum_rec(arr):
    if len(arr)==1:
        return[0]
    else:
        return arr[0] + sum_rec(arr[1:])
lst = [1,2,3,4,5,6]
sum_rec(lst)
```

```
# Regular expression 

import re

help(re)
```

```
# findall  Return a list containing all match 
#search Returns a Match object if there is a match anywhere in the string
#split Returns a list where the string has been split at each match 
#sub Replace one or many matches with a string

#  \w -----> any word character(alpha numeric character) [a-zA-Z0-9]
#  \W -----> any character except word (special characters) [^a-zA-z0-9]

# [] Set of character 
# . Any character

```

```
# Return Empty space

txt = "This is sample program where you can list of words in this text"
x = re.findall("\s",txt)
print(x)
```

```
# Upper case S return all character 

txt = "This is sample program where you can list of words in this text"
x = re.findall("\S",txt)
print(x)
```

```
# Return induvisable word 

txt = "This is sample program where you can list of words in this text"
x = re.findall("\S+",txt)
print(x)
```

```
# Replace character

txt = "How to read the value from HDFS"
x = re.sub("\s","TEST",txt)
print(x)
```

```
# Matching Pattern 

pattern = '^\D+'
test_string = 'python regular 34 exression learn with 22 examples'
result =re.match(pattern , test_string)
if result:
    print("Search successful.", result)
else:
    print("Search unsuccessful" , result)
```
